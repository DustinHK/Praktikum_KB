Memasukan Library Pandas agar dapat membaca file csv atau data 

```python
import pandas as pd
```

untuk membaca data dari file CSV dengan nama "Car_Insurance_Claim.csv" dan di simpan dalam sebuah DataFrame yang dinamai "df"

```python
df = pd.read_csv('Car_Insurance_Claim.csv')
```

1. menghitung rata-rata dari semua kolom df yang bertipe data angka atau('int64')('float64')
mean() untuk menghitung rata2 dari kolom data bertipe angka
Rata2 untuk menyimpan hasil rata2 dari klom data bertipe angka

```python
Rata2 =df.select_dtypes(["int64", "float64"]).mean()
Rata2
```

1.  standar deviasi dari semua kolom df yang bertipe data angka atau ('int64')('float64')
std() menghitung standar deviasi dari masing2 kolom yang bertipe angka
