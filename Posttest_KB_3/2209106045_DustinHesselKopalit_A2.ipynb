##### Memasukan Library seaborn agar dapat membuat visualisasi  csv atau data 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

untuk membaca data dari file CSV dengan nama "Car_Insurance_Claim.csv" dan di simpan dalam sebuah DataFrame yang dinamai "df"

```python
df = pd.read_csv("Car_Insurance_Claim.csv")
```

### PERBANDINGAN JUMLAH PELANGGAN PRIA DAN WANITA

```python
gender_counts = df['GENDER'].value_counts()
sns.set_style('darkgrid')
plt.figure(figsize=(5,7))
plt.bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Pelanggan')
plt.title('Jumlah Pelanggan Berdasarkan Jenis Kelamin')
plt.show()
```

Jenis Kelamin Pelanggan (GENDER): Berdasarkan diagram batang, terlihat bahwa terdapat dua jenis kelamin yaitu "Laki-laki" dan "Perempuan." Jumlah pelanggan Perempuan tampaknya lebih tinggi sedikit dibandingakan Laki-laki.

### DISTRIBUSI  JUMLAH PELANGGAN DENGAN UMUR TERTENTU

```python
plt.hist(df['AGE'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pelanggan')
plt.title('Distribusi Umur Pelanggan')
plt.show()
```

Distribusi Umur Pelanggan (AGE): Dari histogram umur pelanggan, dapat dilihat bahwa distribusi umur cenderung merata, meskipun terdapat beberapa puncak pada rentang tertentu. Mayoritas pelanggan berada pada rentang usia 26-39 TAHUN, sedangkan pelanggan dengan usia 65+ TAHUN lebih sedikit

### PRESENTASE PENGGUNA TIPE KENDARAAN TERTENTU

```python
vehicle_type_counts = df['VEHICLE_TYPE'].value_counts()
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.0f%%', startangle=90)
plt.title('Persentase Tipe Kendaraan')
plt.show()
```

Persentase Tipe Kendaraan (VEHICLE_TYPE): Dari pie chart, dapat dilihat bahwa mayoritas pelanggan memiliki kendaraan tipe Sedan yang mencakup sekitar 95% dari kepemilikan kendaraan di dataset. Sedangkan kendaraan tipe "Sports" hanya mencakup sekitar 5% dari total kepemilikan kendaraan.
Hal ini menunjukkan bahwa kendaraan tipe "Sedan" adalah yang paling umum dimiliki oleh pelanggan dalam dataset ini, sementara kendaraan tipe "Sports" adalah yang paling jarang dimiliki.

### KORELASI ANTAR ATRIBUT NUMERIK

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antara Atribut Numerik')
plt.show()
```

Korelasi Antara Atribut Numerik: Dari heatmap korelasi, terlihat bahwa beberapa atribut numerik memiliki korelasi positif atau negatif. Sebagai contoh, 

1. Korelasi Positif Kuat antara CREDIT_SCORE dan VEHICLE_OWNERSHIP Ini berarti bahwa pelanggan dengan skor kredit yang lebih tinggi cenderung memiliki kepemilikan kendaraan yang lebih tinggi.

2. Korelasi Negatif antara VEHICLE_OWNERSHIP dan OUTCOME Ini mungkin menunjukkan bahwa pelanggan yang memiliki kendaraan sendiri cenderung memiliki tingkat klaim yang lebih rendah atau peringkat risiko yang lebih rendah.
